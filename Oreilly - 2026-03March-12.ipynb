# Recap

Object-oriented programming is a technique for organizing our code -- so that it'll be easier for us to read, write, and maintain over time.  It helps to organize our thoughts, it helps to organize our code, and it helps to maintain a high level of abstraction, so that we're thinking about the big ideas and items we want to work with, rather than low-level ideas such as numbers, lists, tuples, dicts, etc.

1. A "class" is a new data type, one that exists in parallel with strings, lists, tuples, etc. By creating a new, specialized data type, we can solve problems in a clearer way, and think about them more easily, as well. Each class is a new type in Python, and type of noun we can think and talk about.
2. Methods are the verbs in our object-oriented programming world. We define one or more verbs / methods that know how to work with and manipulate our data. Every class will typically have a bunch of methods -- maybe many, maybe only a few.
3. A class is not only a new type of data, but in Python syntax, it's the way that we create a new instance of that class. We invoke the class -- `str()` or `list()` or anything else, such as `Person()`. We invoke the class, and get back a new instance of that type, or of that class.
4. When we create a new instance, Python allocates the memory and does all of the magic behind the scenes. The first we get to influence our object is with the `__init__` method, which is invoked on our object after it is created, but before it's returned to the caller. This gives us a chance to add one or more attributes to our new object. That is the role of `__init__`.
5. The `self` parameter in `__init__` and every other Python method is the instance on which we're running the method. That instance has the attributes that were already defined, and if you want, you can retrieve and/or set attributes in the method.
6. `self` is a parameter, meaning that it's a local variable. It acts like a local variable, and it has the priority of a local variable, because it is one. `self` is a convention, but it's not a keyword and it's not a reserved word.
7. When we assign a new attribute on `self`, that attribute name and value will stick around on the object, even after the method has run. Even after `__init__` is done, the instance on which it ran will retain its attributes.
8. Typically, attributes will be set based on the arguments assigned to parameters.

# Magic methods

We have seen how to implement methods on our objects:

- We write the method
- We invoke the method


There are a number of places in Python where *if* a method by a certain name has been defined, then Python will invoke it for us, without asking.

This is a protocol of sorts:

- We write the method
- Python invokes the method, without even asking us

Some people call these "callbacks."

These methods all start and end with double underscore (`__`), so they are sometimes known as "dunder methods." They are also known as "magic methods."

There are many magic methods, but you only have to define them if you want to modify the default, existing behavior.

One of the many reasons not to define methods that start and end with `__` on your own, without coordinating with the Python documentation, is that you might end up seriously changing the behavior of your object if you choose poorly (or well).

Don't define dunder methods unless you've checked that this name exists, and that it does what you want it to do!

# `len` invokes `__len__`

When you invoke `len` on a Python object, that function looks on the object for the `__len__` method, and invokes it.

- If `__len__` isn't implemented, you get an error.
- If it is, then whatever it returns (must be an integer) is returned from `len`

It's very common for an object that contains others to implement `__len__`.

```python
class Person:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
        self.bank_balance = 0

    def greet(self):
        return f'Hello, {self.first_name} {self.last_name}!'

    def __len__(self):
        return len(self.first_name) + len(self.last_name)

p = Person('Reuven', 'Lerner')        

```

# Two ways to get a string from a value

Python provides *two* ways to get a string from any object value:

- For normal, everyday users, the `__str__` method is used, and returns a string. This is then presented to the user. This is used by `str` and by `print`.
- If you're in Jupyter or a debugger, then you're assumed to be a programmer, and you'll want to see (perhaps) more details. For such people, we have a separate method, `__repr__`, for the *printed representation* of the value.

Which should you implement? 

- Officially, the Python style guide (PEP 8) says that you should use both. `__str__` is used for normal users, and `__repr__` is for internal coding use. Moreover, PEP 8 says that `__repr__` should return a value that could be evaluated as a Python expression.
- If you implement only `__str__`, then it doesn't affect `__repr__` at all.
- If you implement only `__repr__`, it covers for *BOTH* `__str__` and `__repr__`.

My practical advice:
- Only implement `__repr__`
- Have it return whatever you want, and whatever is appropriate
- If, at some point in the future, you want to distinguish between what is shown to regular users, and what is shown to coders, just rename `__repr__` to `__str__`, and implement a new `__repr__`.

# Some other magic methods

Magic methods are everywhere in Python, and they allow you to customize your class. They also allow your class to act like other Python classes. 

- When you compare `a == b`, you're really invoking `a.__eq__(b)`. If you want, you can define `__eq__`, and then determine what makes two objects equal (or not)
- When you retrieve `a[b]`, you're really invoking `a.__getitem__(b)`. If you want, you can implement `__getitem__`, and then return whatever seems appropriate.
- When you add `a + b`, you're really invoking `a.__add__(b)`. If you want, you can implement `__add__` on your class, and then get `+` working for your objects.

This explains why strings, lists, tuples, and dicts all use `[]` to retrieve values. They all implement `__getitem__`.

# A story 

Our company has a `Person` class that we sell to our customers.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f'Hello, {self.name}!'

p1 = Person('name1')
p2 = Person('name2')

print(p1.greet())
print(p2.greet())
```

So far, our company has sold this `Person` class to many customers around the world. 

Some of them would like some additional functionality: They want us to keep track of how many `Person` objects have been created.

How can I do this?

One option: Have a global variable, `population`. Every time we create a new person, `__init__` will update `population` to have 1 more.

```python
population = 0

class Person:
    def __init__(self, name):
        global population   # meaning: don't create a local population; use the global
        self.name = name
        population += 1

    def greet(self):
        return f'Hello, {self.name}!'

print(f'Before, population = {population}')
p1 = Person('name1')
p2 = Person('name2')
print(f'After, population = {population}')

print(p1.greet())
print(p2.greet())
```

I have a global `population`, and it works. But I would like to store this information somewhere else, not as a global.

Here's an idea: 

- `Person` is a class.
- A class in Python is an object.
- Every object has attributes.
- I can add an attribute to any class I want
- Maybe I can add a `population` attribute to my `Person` class, and store things there?

```python
class Person:
    def __init__(self, name):
        self.name = name
        Person.population += 1

    def greet(self):
        return f'Hello, {self.name}!'

# after creating Person, we'll add the population attribute
Person.population = 0


print(f'Before, population = {Person.population}')
p1 = Person('name1')
p2 = Person('name2')
print(f'After, population = {Person.population}')

print(p1.greet())
print(p2.greet())
```

Why am I defining `Person.population` after the class has already been defined? Why don't I just define it *inside* of the class body?

We definitely can do it, but it looks a little weird.

```python
class Person:
    population = 0   # this looks like a variable, but it isn't -- it's an attribute!
    
    def __init__(self, name):
        self.name = name
        Person.population += 1

    def greet(self):
        return f'Hello, {self.name}!'

# as soon as Person has been defined, it has *three* class
# attributes -- population, __init__, and greet

# yes, methods are attributes on the class!

print(f'Before, population = {Person.population}')
p1 = Person('name1')
p2 = Person('name2')
print(f'After, population = {Person.population}')

print(p1.greet())
print(p2.greet())
```

# Attribute lookup is the key to understanding objects

- Python asks for `p1.name`. Does `p1` have an attribute `name`? Yes. We get the value.
- Python asks for `Person.population`. Does `Person` have an attribute `population`? Yes. We get the value.
- Python asks for `p1.population`. Does `p1` have an attribute `population`? NO.
    - Python doesn't give up, and turns to the class of `p1`, and asks: Does `Person` have `population`? The answer is "yes."
 
Neither `p1` nor `p2` has a `population` attribute. But if you ask for it, and if there isn't any attribute with that name on the instance, Python will look on the class.

You knew this already! Beause when you invoke a method, you invoke it on the instance, but it is stored as an attribute on the class!

This means that we can always retrieve a class attribute (whether data or a function) via the instance. Maybe it'll take a bit longer, but Python will get there as part of its attribute search + lookup.

You can even, in a method, say `self.population`, which will look on the instance, not find it, and then look on the class. That is *TOTALLY FINE*. HOWEVER, don't assign to it via `self`, because that will create a new attribute on the *instance*! Then you'll have the same name for attributes on the instance and the class, which is a recipe for confusion + disaster.

# Part of the ICPO lookup

When we ask for an attribute on an object in Python:

- `I` -- instance: First, we turn to the instance, the original value on which the request is made
- `C` -- class: If the attribute is not there, we go to that value's class

# Inheritance!

We've spoken about *composition*, the idea that one object can contain another. This is a relationship we can describe as a "has-a" relationship. One object has-a nother.

Inheritance is a different type of relationship. It's an "is-a" relationship. Inheritance is all about reusing code, and avoiding the need to reinvent the wheel. 

If a class already exists, and if your new class is basically like that one, but with some adjustments, then why write an entirely new one? You can "borrow" the parts that you like, and write the new things that are different from it.

This is inheritance.

- Inheritance is *not* the only way to write objects
- If your class is basically the same as an existing one, then by all means use inheritance! But you don't have to.

```python
# another version of the previous story

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f'Hello, {self.name}!'

p1 = Person('name1')        
p2 = Person('name2')

print(p1.greet())
print(p2.greet())
```

# Inheritance 

It's basically DRY -- don't repeat yourself!

Why re-implement the same code, when you can just use an existing class and make minor modifications?

Inheritance works in Python by extending the way in which attributes are searched for.

If before we had:

- Instance -- first look on the object that was named in the expression
- Class -- then, if not on the instance, search on the class

Now we also have 

- Parent -- if not on the class, search on the class's parent class

```python
# here is our Person - Employee class hierarchy finally working (sort of)

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f'Hello, {self.name}!'

p1 = Person('name1')        
p2 = Person('name2')

print(p1.greet())
print(p2.greet())


class Employee(Person):  # this means: Employee is-a Person, inherits from Person
    def __init__(self, name, id_number):
        self.name = name
        self.id_number = id_number

    def greet(self):
        return f'Hello, {self.name}!'

e1 = Employee('emp1', 1)        
e2 = Employee('emp2', 2)

print(e1.greet())
print(e2.greet())
```

Because there were no missing attributes in the child class, we never had to go up to `Person` to resolve any attributes. 

What if we get rid of `Employee.greet`? It's identical to `Person.greet`.

```python
# here is our Person - Employee class hierarchy finally working (sort of)

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f'Hello, {self.name}!'

p1 = Person('name1')        
p2 = Person('name2')

print(p1.greet())
print(p2.greet())


class Employee(Person):  # this means: Employee is-a Person, inherits from Person
    def __init__(self, name, id_number):
        self.name = name
        self.id_number = id_number

e1 = Employee('emp1', 1)        
e2 = Employee('emp2', 2)

print(e1.greet()) # Does e1 have greet? No. Does Employee have greet? No. Does Person have greet? Yes
print(e2.greet())
```

What happened here? The lookup that we discussed already.

Python wanted `self.name` inside of `Person.greet`:

- It turned to `self`, an instance of `Employee`. Does it have `name`? No.

Why not?

It all comes down to `__init__`.

# Three paradigms for method inheritance

- The child class does nothing, and lets the parent method execute. That's what we did with `Employee.greet` -- we got rid of it. Now, `Person.greet` runs for instances of `Person` and also instances of `Employee`.
- The child class implements a method, and never calls the method in the parent class. This is great if the child class will *override* the parent class's implementation.
- The child class wants to do its own thing but *also* wants the parent implementation. We want to combine them somehow. For this to happen, we need to call the parent method from the child method.

How can we invoke the parent class's method from within the child class's method?

```python
# we'll just call Person.__init__ from Employee.__init__

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f'Hello, {self.name}!'

p1 = Person('name1')        
p2 = Person('name2')

print(p1.greet())
print(p2.greet())


class Employee(Person):  # this means: Employee is-a Person, inherits from Person
    def __init__(self, name, id_number):
        Person.__init__(self, name) # this will set name on self!
        self.id_number = id_number

e1 = Employee('emp1', 1)        
e2 = Employee('emp2', 2)

print(e1.greet()) # Does e1 have greet? No. Does Employee have greet? No. Does Person have greet? Yes
print(e2.greet())
```

```python
# the other way to do things is with the super() builtin
# in some languages, super() invokes the method of the same name in the parent
# that's *not* how it works in Python.

# here, super() is a finder, it looks for the class where the method should be found

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f'Hello, {self.name}!'

p1 = Person('name1')        
p2 = Person('name2')

print(p1.greet())
print(p2.greet())


class Employee(Person):  # this means: Employee is-a Person, inherits from Person
    def __init__(self, name, id_number):
        super().__init__(name)   # notice that self is passed automatically
        self.id_number = id_number

e1 = Employee('emp1', 1)        
e2 = Employee('emp2', 2)

print(e1.greet()) # Does e1 have greet? No. Does Employee have greet? No. Does Person have greet? Yes
print(e2.greet())
```

# The full attribute search path

When Python looks for attributes, it looks in 4(ish) places:

- I -- the instance
- C -- the instance's class
- P* -- the parent of the instance's class. A class might inherit from another class, which inherits from another class... Python will keep following the parents all the way, until someone eventually inherits from ... `object`
- O -- `object`, the top of Python's object hierarchy
