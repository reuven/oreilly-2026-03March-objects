# What are objects?

Software has always been hard to write, and even harder to maintain.

In the 1970s, people started to talk about how to write software in more manageable ways. Alan Kay at Xerox PARC -- he invented a new idea/paradigm for writing software, called object-oriented programming.

He decided to model software on biological systems, such as the human body. Our bodies are made up of many cells. Each cell has a type. Each type of cell receives certain messages and sends certain messages. Just as the body can have many small parts, each with a type, and each sending/receiving certain types of messages, maybe we can create our software in this way, and thus think about it as a collection of objects, each having a specific type.

Kay wasn't talking about a technological change for faster execution, or for less memory usage. He was talking about a change for easier management. Easier debugging. Easier creation and refactoring of software. 

To translate from biology-speak to programming-speak:

- Every cell is an object.
- Every type of cell is a class.
- Instead of sending/receiving messages, we talk about invoking methods.



# Everything is an object

Who cares? So what if everything in Python is an object?

It means several things:

- we can apply the same rules to data structures that we create as the system itself uses on its own data structures.
- we can extend/improve the system by simply writing new classes (with data structures and methods)

There is standardized way to do things in general, and that standardization extends to extending the langauge with new nouns and verbs.

once you understand how Python objects work, the core language makes a lot more sense, too!
