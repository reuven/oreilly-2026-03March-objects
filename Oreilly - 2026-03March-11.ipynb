# What are objects?

Software has always been hard to write, and even harder to maintain.

In the 1970s, people started to talk about how to write software in more manageable ways. Alan Kay at Xerox PARC -- he invented a new idea/paradigm for writing software, called object-oriented programming.

He decided to model software on biological systems, such as the human body. Our bodies are made up of many cells. Each cell has a type. Each type of cell receives certain messages and sends certain messages. Just as the body can have many small parts, each with a type, and each sending/receiving certain types of messages, maybe we can create our software in this way, and thus think about it as a collection of objects, each having a specific type.

Kay wasn't talking about a technological change for faster execution, or for less memory usage. He was talking about a change for easier management. Easier debugging. Easier creation and refactoring of software. 

To translate from biology-speak to programming-speak:

- Every cell is an object.
- Every type of cell is a class.
- Instead of sending/receiving messages, we talk about invoking methods.



# Everything is an object

Who cares? So what if everything in Python is an object?

It means several things:

- we can apply the same rules to data structures that we create as the system itself uses on its own data structures.
- we can extend/improve the system by simply writing new classes (with data structures and methods)

There is standardized way to do things in general, and that standardization extends to extending the langauge with new nouns and verbs.

once you understand how Python objects work, the core language makes a lot more sense, too!

# Jargon/vocabulary 

- `class` and `type` -- in Python, these are basically interchangeable when we talk about things, although they're used in different places when we're actually coding. This is a category of object, of value. You've seen classes before -- strings are the `str` class, and dictionaries are the `dict` class. The point of object-oriented programming is to create your *own* classes, above and beyond what Python has created and provides.
    - When we talk about a string, such as `'abcd'`, we're really referring to an object whose class is `str`, or string.
    - That string is known as an *instance* of `str`. Or sometimes as an *object* of type `str`.
    - Python has millions of objects running around, and every single one of them is an instance of some class. The class defines the object's behavior.
- If I want to create a new value of type `X` (that is, `X` is the type), then I invoke `X()`. That gives me a new value of type `X` back.
    - I can create a new integer value, aka an instance of `int`, with `int()`.
    - I can create a new string value, an instance of `str`, by invoking `str()`.
    - Similarly, I can create a new list with `list()`.
- `instance` -- this word basically means, "A value whose class is ___". You'll say, "This is an instance of `str`," or "This is an instance of `dict`."
- `object` -- This is used too much. On the one hand, it just means a value. Every value in Python is an object, so that's not wrong. Sadly, `object` is also the name of a class in Python -- the top of the hierarchy, from which everyone inherits. 

# How does it matter what a particular value's class is?

This actually determines *everything* about a value!

A class determines lots of things about its instances:

- What kinds of data it stores
- How you can retrieve the data
- What methods you can invoke on that data
- What error messages you'll get if/when things go wrong

Once you know the type of an object, you know what you can (and cannot) do with it.



# Do we really need objects?

No. We don't. But they make things much easier and handier.

What if I want to replicate objects, but not using objects? In Python, we typically use tuples for this sort of thing.

```python
# here, I'm creating a tuple with 3 elements representing first name, last name, and shoe size.
# there is nothing wrong with this!

p = ('Reuven', 'Lerner', 46)
```

```python
p[0]
```

```python
p[1]
```

```python
p[2]
```

```python
# If I want to print  both first and last names, I can!

print(f'{p[0]} {p[1]}')
```

```python
# I could even write a function that returns one string, the full name, based on this tuple

def fullname(person_tuple):
    return f'{person_tuple[0]} {person_tuple[1]}'
```

```python
fullname(p)
```

# This worked, but...

1. We lost the advantage of *abstraction*.
2. There's no guarantee that `p` will be initialized with the right data in the right elements. `fullname` expects what we just did, but there's no enforcement.
3. I hate using integers to retrieve values from a data structure.
