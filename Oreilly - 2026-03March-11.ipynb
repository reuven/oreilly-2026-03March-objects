# What are objects?

Software has always been hard to write, and even harder to maintain.

In the 1970s, people started to talk about how to write software in more manageable ways. Alan Kay at Xerox PARC -- he invented a new idea/paradigm for writing software, called object-oriented programming.

He decided to model software on biological systems, such as the human body. Our bodies are made up of many cells. Each cell has a type. Each type of cell receives certain messages and sends certain messages. Just as the body can have many small parts, each with a type, and each sending/receiving certain types of messages, maybe we can create our software in this way, and thus think about it as a collection of objects, each having a specific type.

Kay wasn't talking about a technological change for faster execution, or for less memory usage. He was talking about a change for easier management. Easier debugging. Easier creation and refactoring of software. 

To translate from biology-speak to programming-speak:

- Every cell is an object.
- Every type of cell is a class.
- Instead of sending/receiving messages, we talk about invoking methods.



# Everything is an object

Who cares? So what if everything in Python is an object?

It means several things:

- we can apply the same rules to data structures that we create as the system itself uses on its own data structures.
- we can extend/improve the system by simply writing new classes (with data structures and methods)

There is standardized way to do things in general, and that standardization extends to extending the langauge with new nouns and verbs.

once you understand how Python objects work, the core language makes a lot more sense, too!

# Jargon/vocabulary 

- `class` and `type` -- in Python, these are basically interchangeable when we talk about things, although they're used in different places when we're actually coding. This is a category of object, of value. You've seen classes before -- strings are the `str` class, and dictionaries are the `dict` class. The point of object-oriented programming is to create your *own* classes, above and beyond what Python has created and provides.
    - When we talk about a string, such as `'abcd'`, we're really referring to an object whose class is `str`, or string.
    - That string is known as an *instance* of `str`. Or sometimes as an *object* of type `str`.
    - Python has millions of objects running around, and every single one of them is an instance of some class. The class defines the object's behavior.
- If I want to create a new value of type `X` (that is, `X` is the type), then I invoke `X()`. That gives me a new value of type `X` back.
    - I can create a new integer value, aka an instance of `int`, with `int()`.
    - I can create a new string value, an instance of `str`, by invoking `str()`.
    - Similarly, I can create a new list with `list()`.
- `instance` -- this word basically means, "A value whose class is ___". You'll say, "This is an instance of `str`," or "This is an instance of `dict`."
- `object` -- This is used too much. On the one hand, it just means a value. Every value in Python is an object, so that's not wrong. Sadly, `object` is also the name of a class in Python -- the top of the hierarchy, from which everyone inherits.

- What error messages you'll get if/when things go wrong

Once you know the type of an object, you know what you can (and cannot) do with it.

# What's going on here?

- We use `class` to tell Python that we're creating a new data structure. We're teaching Python what a `Person` is.
- We name our class, using "CamelCase."
- We have a colon at the end of that line
- Then we have an indented block. This is the body of the class definition. This is where we tell Python what's special/different about our new class.
- Inside of the class body, we can define functions. But when you define a function (with `def`) inside of a class, you actually define a *method*! Because we're doing it inside of a class, Python knows to define it as a method, rather than a regular function.
- The only method we've defined here is `__init__`, which we never (or almost never) invoke directly. Rather, we'll let Python invoke it on our behalf after the new instance of `Person` has been creatd.
- The new instance of `Person`, once created behind the scenes, is passed to `__init__` as the first argument. It is thus assigned to `self`, the first parameter in our method.

# Attributes 

In Python, we have *two* places where we can store things:

- Variables
- Attributes

Variables are easy to understand. We define a variable and assign to it.

Attributes are like a private dictionary on every value in Python. Every integer, string, dict, etc. has its own private dict, known as attributes, with names and values. The syntax doesn't look like a dict; there are no `[]`.  But it's name-value pairs regardless.

Every time you see Python code that looks like

    a.b

this means:

- Go to `a`, whether it's a variable or a value
- Get the value, if it's not one already
- Ask it whether it has a `b` attribute (i.e., whether the key "b" exists)
- If so, get the value for that attribute

When we say `self.name = name`, we're basically saying:

- Get the value of the `name` parameter (a variable)
- Assign that value to a attribute on `self`, the new instance of Person
- Once we've done that, we can always retrieve and use the `name` attribute on an instance of `Person`
- The fact that we don't have to invoke `__init__` ourselves is great -- it means that we are guaranteed that every instance of `Person` has a name.

````python
# let's fill out our Person class

class Person:
    def __init__(self, first_name, last_name, shoe_size):
        self.first_name = first_name
        self.last_name = last_name
        self.shoe_size = shoe_size

p = Person('Reuven', 'Lerner', 46)        
````

````python
p.first_name
````

````python
p.last_name
````

````python
p.shoe_size
````

# What happened in `__init__`?

1. I invoked `Person`, giving it `'Reuven'` (and other values) as arguments
2. Python created a new instance of `Person`. This is done in `__new__`, the "constructor" method that you will almost certainly never have to write, and can rely on the default.
3. That new instance of `Person` is passed to `__init__`, along with any arguments that were passed to `Person`.
4. `__init__` assigns each of the parameters (other than `self`) to be attributes on `self`

You don't need to set all attributes on an object in `__init__`. But it's a very good idea to do so, because it makes reading, understanding, and maintaining your class much easier.

# What the heck is `self`?

First, it's not a keyword. It's not a reserved word. It is a very, *very* strong convention in the Python world.

The idea is: `self` is the variable that refers to the current instance. So when we invoke a method, that method is invoked on an instance, and that instance is available via `self`.

If I say `s.upper()`, and `s` is a string, then `str.upper` gets `s` as a value for its `self`.

Why not use a word like `this`?  (Tradition! And it comes from SmallTalk.)

Why is `self` a parameter? Why not just make it a special keyword? If it were special, then we would need special rules for it. But because it's a parameter, it just follows the same rules as all other local variables and parameters.

# We're msising methods!

(Aside from `__init__`)

How can create methods? Why would we want to create methods?

Let's start with the most standard, expected methods that people want to write: getters and setters

To write a method:
1. Write a function inside of the class body
2. Make sure that you make `self` the first parameter, which will get the instance
3. Methods have *all* of the same facilities and tricks and techniques as regular functions do. 

````python
class Person:
    def __init__(self, name):
        self.name = name

    def get_name(self):
        return self.name   # notice -- I must use self to retrieve the name attribute

    def set_name(self, new_name):
        self.name = new_name

p = Person('Reuven')        
print(p.get_name())

p.set_name('newReuven')
print(p.get_name())
````

# This works great!

However: We don't normally write getter and setter methods in Python.

In Python, we have direct access to all of the attributes. (There is no sense of "protected" or "private" as other languages have.)

We want to retrieve an attribute? We retrieve it!

We want to set an attribute? We set it!

It's rare to have getters and setters.

````python
class Person:
    def __init__(self, name):
        self.name = name

p = Person('Reuven')        
print(p.name)  # just grab the attribute

p.name = 'newReuven'
print(p.name)
````

# Do we not write methods, then?

We do write methods! But we just write them to do things other than getters and setters. For example, I could have a `fullname` method on my `Person` class.

````python
class Person:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    def fullname(self):
        return f'{self.first_name} {self.last_name}'

p = Person('Reuven', 'Lerner')        
print(p.fullname())
````

# No private?

Technically speaking, there are *no* private attributes in Python. Everything is 100% public.

However, if you want an attribute to be seen as private, you can start its name with an `_`. That doesn't change anything technically, but it does signal to other people that this should be treated as private, and/or shouldn't be touched.

We have come full circle!

- We have our own data structure for people, and a way to get the full name, just as we did before with our tuple
- However, now we can be sure that the data structure on which we run `full_name`, actually has all of the attributes we need.
- It also uses the names, which makes it easier to read/write/understand.

If you don't have `self` as the first parameter, that's (kind of) OK -- Python will simply assign the instance to whatever the first parameter is. You're likely to get an error, because your code was probably not written on the assumption that the first parameter would get the instance.

# Composition of objects

I find that people are always looking for excuses (reasons) to use inheritance when writing code with objects. Inheritance, as we'll discuss tomorrow, is for one particular use case: You're writing a class, and its implementation overlaps greatly with another class. For that reason, you can "borrow" the other class's implementation except for where you want to be different.

A very large number of object-oriented problems have nothing to do with this. Inheritance is nice, but it's not used everywhere.

But what is used everywhere is *composition*, meaning one object owning another. If you can say that one object has-a(nother) object, then that's as composition relationship.

In Python, we use composition *all* of the time. If an attribute in an object contains a string or an integer, that's composition - one object holds another (because strings and numbers are both objects).



# How do you know what to invoke?

So far, I've been telling you what methods to write, and also what methods to invoke. But in the real world, how can you know what methods are even available on a class, let alone how to invoke them?

It all comes down to "docstrings," the strings (typically triple quoted) that we put at the top of a function -- we can also put them at the top of a class or a method!

````python
help(Bowl) # this will display the docstring for Bowl, and everything in it
````

````python
help(Person)
````

````python
class Person:
    """A class to represent people. Each person has a first and last name.

    Each instance of Person has two attributes:
    - first_name -- a string, the person's first name
    - last_name -- a string, the person's last name
    """
    
    def __init__(self, first_name, last_name):
        """initializer method for Person

        - Requires: two strings, for first_name and last_name
        - Modifies: the instance, adding two attributes
        - Returns: None
        
        """

        self.first_name = first_name
        self.last_name = last_name

    def fullname(self):
        """get a full name from the Person
        """
        return f'{self.first_name} {self.last_name}'

p = Person('Reuven', 'Lerner')        
print(p.fullname())
````
